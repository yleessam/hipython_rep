# 넘파이의 사용 

1. 데이터 분석 프로젝트 : 수치형 데이터의 요약, 통계, 시각화
2. 이미지 처리 프로젝트 : 이미지 분석
3. 게임, 시뮬레이션 프로그램 : 수치시뮬레이션 등을 구현
4. 머신러닝/딥러닝 모델 프로젝트 : 금융데이터, 의료데이터, 실험데이터, 센서데이터

```python
import numpy as np 

data = [
    [120, 135, 150, 145],  # 제품A
    [100, 110, 140, 130],  # 제품B
    [90, 100, 95, 105] ]

data
```

```python
sales_data = np.array(data)
np.sum(sales_data, axis=1) #총판매량
```

```python
profit_per_unit = np.array([20,30,25])
total_sales = np.dot(profit_per_unit, sales_data)
total_sales #총판매금액
```

## 데이터 시뮬레이션

1차원 배열의 랜덤워크 시뮬레이션

```python
import matplotlib.pyplot as plt

n_step = 100000
step = np.random.choice([-1,1], size=n_step)
print(step)
walk = np.cumsum(step)

plt.plot(walk)
```

## 슬라이싱

- [:]  전체
- [0:n] 0~n-1까지
- [:5] 0~4까지

리스트와 차이점
- 리스트 : 복사본을 리턴
- 배열 : 원본 배열의 뷰 -> 복사x, 원본에도 영향을 미친다.
- 슬라이싱.copy()


```python
x = np.arange(10)
x
```

```python
x[:5]
```

```python
x[4:7]
```

```python
new_x = x[4:7]
new_x
```

```python
new_x[0]=100
new_x, x  #view , 뷰의 값 변경 -> 원본의 값도 변경
```

```python
new_x1 = x[4:7].copy()
new_x1[0]=200
new_x1, x
```
