```python
import time
points = {}
# # ë©”ë‰´ ì €ì¥ ë¦¬ìŠ¤íŠ¸ -> ë©”ë‰´ ì¶”ê°€ ì‹œ ì—¬ê¸°ì„œ.
menu = [
    {'name':'ìƒ¤í”„', 'price': 6000},
    {'name':'ë³¼íœ', 'price': 3000},
    {'name':'ì§€ìš°ê°œ', 'price': 1000},
    {'name':'í™”ì´íŠ¸', 'price': 3000},
    {'name':'ê³µì±…', 'price': 3500}
]

def prt_menu(menu):
    print('â•”' + 'â•' * 40 + 'â•—')
    print(f'ë¹µê¾¸ë˜¥ê¾¸ ë¬¸êµ¬ì•¼'.center(34))
    print(f"ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!ğŸ’¥".center(34))
    print('â•š' + 'â•' * 40 + 'â•') 
    print(f'{"ë²ˆí˜¸"}\t\t{"ì œí’ˆëª…"}\t\t{"ê°€ê²©"}')
    print('-'*44)

    for i, m in enumerate(menu):
        print(f'{i+1}\t\t{m["name"]}\t\t{m["price"]:,}ì›')
    print('-'*44)

def prt_alarm():
    print('='*44)
    print(f'{"[ì•Œë¦¼]âŒ ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤.":^35}')
    print(f'{"[ì•Œë¦¼]âœ… ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.":^35}')
    print('='*44)

def prt_time():
    for i in range(3, 0, -1):
        print('=' * 44)
        print(f'{i}ì´ˆ ë’¤ì— ì‹œì‘í™”ë©´ìœ¼ë¡œ ëŒì•„ê°‘ë‹ˆë‹¤.'.center(34))
        time.sleep(1)
        
def add_ord():
    while True:        
        try:
            ord_num = int(input('ì£¼ë¬¸í•˜ì‹¤ ì œí’ˆ ë²ˆí˜¸ë¥¼ ì„ íƒí•˜ì„¸ìš”: ')) -1
            if not (0 <= ord_num < len(menu)):
                prt_alarm()
                continue
            return ord_num
        except ValueError:
            prt_alarm()
            continue
        
def add_qty():
    while True:
        qty = input('\tìˆ˜ëŸ‰ì„ ì…ë ¥í•˜ì„¸ìš”(ì·¨ì†Œ: x): ')
        if qty.lower() =="x":
            return qty
        try:
            qty = int(qty)
            if qty > 0:
                return qty
            prt_alarm()
        except ValueError:
            prt_alarm()

def add_lst(order_list, ord_num, qty):
    item = menu[ord_num]
    for i in order_list:
        if i['name'] == item['name']:
            i['qty'] += qty
            return order_list 
    order_list.append({
        'name': item['name'],
        'qty': qty,
        'price': item['price']
    })
    return order_list
    
def yesno():
    while True:
        a=input('ë” ì£¼ë¬¸í•˜ì‹œê² ìŠµë‹ˆê¹Œ?(y/n):')
        if a.lower() not in ['y','n']:
            prt_alarm()
            continue
        else:
            return a
            
def calc_cost():
    total = 0   
    print('='*44)
    print(f'\t\t{"ì£¼ë¬¸ ë‚´ì—­"}')
    print('-'*44)

    for order in order_list:
        item_total = order['qty'] * order['price']
        total += item_total

        print(f'{order["name"]}\t{order["qty"]}ê°œ\t-\t{item_total:,}ì›')

    print('-'*44)
    print(f'ì´ ê²°ì œ ê¸ˆì•¡: {total:,}ì›')
    print('='*44)
    return total

def prt_points():
    while True:
        ab = input('í¬ì¸íŠ¸ë¥¼ ì ë¦½í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ')
        if ab.lower() not in ['y', 'n']:
            prt_alarm()
            continue
        elif ab.lower()=='n':
            return        
        else:
            while True:
                phone = input('ì „í™”ë²ˆí˜¸ë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš”: ')
                if not phone.isdigit() or len(phone) < 11 or phone == "x":
                    prt_alarm()
                    continue                       
                user_id = f'user_{phone[-4:]}'
                point = int(total * 0.05)                 
                if user_id in points :
                    points[user_id] += point
                else :
                    points[user_id] = point
                print('='*44)
                print(f'ğŸ{user_id}ê³ ê°ë‹˜, {point}ì  ì ë¦½ ë˜ì—ˆìŠµë‹ˆë‹¤!ğŸ')
                print(f' ê³ ê°ë‹˜ì˜ ì´ ì ë¦½í¬ì¸íŠ¸ëŠ” {points[user_id]:,}ì  ì…ë‹ˆë‹¤!')
                print('='*44)
                return
     
######################################################################################################################################################


def draw_pencil_ascii():
    pencil_art = r"""
        .================ooooooo
 ___   ,'    \_________________________________
/   /-/       /                   ////////////  ''--..._
\___\-\       \                   \\\\\\\\\\\\  __..--'
       `--------------------------''''''''''''''
"""
    print(pencil_art)

draw_pencil_ascii()
while True:
    prt_menu(menu)
    # ë‹¤ì¤‘ ì£¼ë¬¸ ì €ì¥ ë¦¬ìŠ¤íŠ¸
    order_list = []

    while True: 
        
        ord_num = add_ord()         
        menut = menu[ord_num]['name']
        
        print('='*44)
        print(f'{"ê³ ê°ë‹˜ì´ ì„ íƒí•˜ì‹  ì œí’ˆì€":^33}')
        print(f'{menut + " ì…ë‹ˆë‹¤.":^33}')
        print(f'{"ëª‡ ê°œë¥¼ ë“œë¦´ê¹Œìš”?":^33}')
        
        
        qty = add_qty()
        if qty in ('x','X'):
            break
             
        add_lst(order_list, ord_num, qty)
        
        answer = yesno()
        if answer == 'n':
            break
        print('-'*44)

    if qty not in ('x','X'):
        
        total = calc_cost()
        
        prt_points()
        
        print(f'{"ê²°ì œê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.":^37}')
        print(f'{"ì ì‹œë§Œ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”. ğŸ˜Š":^37}')

        prt_time()
```
